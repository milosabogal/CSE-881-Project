```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
from matplotlib.colors import ListedColormap

colors = ['blue', 'red']
cmap = ListedColormap(colors)

n_splits = 5
split_size = 100 // n_splits
train_size = 0.9

for i in range(n_splits):
    dataset = [0]*100
    dataset[i*split_size:(i+1)*split_size] = [1]*split_size
    plt.scatter(range(100), [i]*100, c=dataset, marker='_', lw=10, cmap=cmap)

plt.xlim([0, 100])
plt.ylim([n_splits-0.5, -0.5])
plt.yticks(np.arange(n_splits))
plt.xlabel('Time')
plt.ylabel('CV Fold')
plt.title("K-fold Cross-Validation")
plt.legend([Patch(color='blue'), Patch(color='red')], ['Training set', 'Testing set']);
```

```python
n_splits = 5
split_size = 100 // n_splits
train_size = 0.9

dataset = [0]*int(train_size*split_size) + [1]*int(split_size - int(train_size*split_size))
for i in range(n_splits):
    plt.scatter(range(i*split_size,(i+1)*split_size), [i]*split_size, c=dataset, marker='_', lw=10, cmap=cmap)

plt.xlim([0, 100])
plt.ylim([n_splits-0.5, -0.5])
plt.yticks(np.arange(n_splits))
plt.xlabel('Time')
plt.ylabel('CV Fold')
plt.title("Blocke Time Series Cross-Validation")
plt.legend([Patch(color='blue'), Patch(color='red')], ['Training set  (90%)', 'Testing set (10%)']);
```
